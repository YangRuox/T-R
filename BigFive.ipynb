```python
import streamlit as st

# BFI-44 questions list
questions = [
    "I am the life of the party.",  # Extraversion
    "I don't talk a lot.",          # Extraversion
    "I feel comfortable around people.",  # Extraversion
    "I keep in the background.",   # Extraversion
    "I start conversations.",      # Extraversion
    "I have little to say.",       # Extraversion
    "I talk to a lot of different people at parties.", # Extraversion
    "I don't like to draw attention to myself.",  # Extraversion
    "I am quiet around strangers.",  # Extraversion
    "I enjoy trying new things.",   # Openness
    "I have a rich vocabulary.",    # Openness
    "I have a vivid imagination.", # Openness
    "I am quick to understand things.", # Openness
    "I spend time reflecting on things.", # Openness
    "I am not interested in abstract ideas.",  # Openness
    "I do not like art.",           # Openness
    "I have difficulty understanding abstract ideas.", # Openness
    "I have a lot of artistic interests.",  # Openness
    "I am easily disturbed.",      # Neuroticism
    "I get upset easily.",         # Neuroticism
    "I change my mood a lot.",     # Neuroticism
    "I get nervous easily.",       # Neuroticism
    "I worry about things.",       # Neuroticism
    "I often feel blue.",          # Neuroticism
    "I am relaxed most of the time.",  # Neuroticism
    "I rarely feel depressed.",    # Neuroticism
    "I am easily embarrassed.",    # Neuroticism
    "I feel little concern for others.",  # Agreeableness
    "I am not interested in other people's problems.",  # Agreeableness
    "I insult people.",            # Agreeableness
    "I sympathize with others' feelings.", # Agreeableness
    "I take time out for others.", # Agreeableness
    "I am not really interested in others.", # Agreeableness
    "I feel others' emotions.",    # Agreeableness
    "I make people feel at ease.",  # Agreeableness
    "I am often preoccupied with details.",  # Conscientiousness
    "I follow a schedule.",        # Conscientiousness
    "I am exacting in my work.",   # Conscientiousness
    "I am always prepared.",       # Conscientiousness
    "I leave my belongings around.",  # Conscientiousness
    "I pay attention to details.",  # Conscientiousness
    "I make plans and stick to them.",  # Conscientiousness
    "I get chores done right away.", # Conscientiousness
]

# Reverse scoring items
reverse_items = [1, 3, 5, 6, 7, 8, 10, 11, 13, 14, 15, 16, 18, 19, 20, 22, 24, 25, 27, 28, 30, 31, 32, 33, 35, 37, 38, 39, 40, 41]

# Create a Streamlit app to display the questions and collect answers
def run_bfi_app():
    st.title("Big Five Inventory (BFI-44) Personality Questionnaire")

    # List to store answers
    answers = []

    # Display questions with radio buttons for answers
    for idx, question in enumerate(questions):
        answer = st.radio(question, [1, 2, 3, 4, 5], key=f"q{idx}")
        answers.append(answer)

    # Submit button
    if st.button("Submit Answers"):
        if len(answers) == 44:
            # Calculate scores
            scores = [0, 0, 0, 0, 0]  # Corresponding to Big Five: E, O, N, A, C
            for i, ans in enumerate(answers):
                if i < 8:  # Extraversion (E)
                    scores[0] += ans if i not in reverse_items else (6 - ans)
                elif i < 16:  # Openness (O)
                    scores[1] += ans if i not in reverse_items else (6 - ans)
                elif i < 24:  # Neuroticism (N)
                    scores[2] += ans if i not in reverse_items else (6 - ans)
                elif i < 32:  # Agreeableness (A)
                    scores[3] += ans if i not in reverse_items else (6 - ans)
                else:  # Conscientiousness (C)
                    scores[4] += ans if i not in reverse_items else (6 - ans)

            # Display the scores
            st.subheader("Your Big Five Personality Scores:")
            st.write(f"Extraversion (E): {scores[0]}")
            st.write(f"Openness (O): {scores[1]}")
            st.write(f"Neuroticism (N): {scores[2]}")
            st.write(f"Agreeableness (A): {scores[3]}")
            st.write(f"Conscientiousness (C): {scores[4]}")

            # Provide a brief analysis
            st.subheader("Personality Analysis:")
            st.write("Based on your answers, your personality scores are as follows:")
            if scores[0] > 32:
                st.write("You score high on Extraversion, which suggests you enjoy social interactions and seek out excitement.")
            else:
                st.write("You score low on Extraversion, which suggests you may prefer quiet, introspective activities.")

            if scores[1] > 32:
                st.write("You score high on Openness, indicating you enjoy new experiences and are open-minded.")
            else:
                st.write("You score low on Openness, which suggests you may prefer conventional and practical ideas.")

            if scores[2] > 32:
                st.write("You score high on Neuroticism, meaning you may experience more anxiety and emotional instability.")
            else:
                st.write("You score low on Neuroticism, suggesting you tend to be more calm and stable under pressure.")

            if scores[3] > 32:
                st.write("You score high on Agreeableness, meaning you care about others' feelings and enjoy helping others.")
            else:
                st.write("You score low on Agreeableness, suggesting you may be more independent or less empathetic.")

            if scores[4] > 32:
                st.write("You score high on Conscientiousness, meaning you are organized, dependable, and goal-oriented.")
            else:
                st.write("You score low on Conscientiousness, suggesting you may be more flexible and spontaneous.")

        else:
            st.warning("Please answer all the questions!")

# Run the app
if __name__ == "__main__":
    run_bfi_app()
```

```python
# Êñá‰ª∂ÂêçÔºöbigfive_app.py
import streamlit as st
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler
from joblib import load

# Âä†ËΩΩËÆ≠ÁªÉÂ•ΩÁöÑÊ®°ÂûãÂíåÊï∞ÊçÆ
model = torch.load("your_model.pth", map_location=torch.device("cpu"))
model.eval()

scaler = load("your_scaler.pkl")  # ‰Ω†ËÆ≠ÁªÉÊ®°ÂûãÊó∂‰ΩøÁî®ÁöÑ scaler
job_names = np.load("job_names.npy")  # 263‰∏™ËÅå‰∏öÂêçÁß∞
job_codes = np.load("job_codes.npy")  # ÂØπÂ∫îËÅå‰∏ö‰ª£Á†Å
scaled_features = np.load("scaled_features.npy")  # ÂéüÂßãËÅå‰∏öÁâπÂæÅÔºàÁî®‰∫éÁõ∏‰ººÂ∫¶Ôºâ
similarity_matrix = np.load("similarity_matrix.npy")

# 44ÈÅìÈ¢ò + ÊØèÈ¢òÊâÄÂ±ûÁª¥Â∫¶ + Ê≠£ÂèçÂêë
items = [
    # (È¢òÁõÆ, Áª¥Â∫¶Âêç, ÊòØÂê¶ÂèçÂêë)
    ("I am the life of the party.", "Extraversion", False),
    ("I don't talk a lot.", "Extraversion", True),
    ("I feel comfortable around people.", "Extraversion", False),
    ("I keep in the background.", "Extraversion", True),
    ("I start conversations.", "Extraversion", False),
    ("I have little to say.", "Extraversion", True),
    ("I talk to a lot of different people at parties.", "Extraversion", False),
    ("I don't like to draw attention to myself.", "Extraversion", True),
    ("I am quiet around strangers.", "Extraversion", True),
    
    ("I enjoy trying new things.", "Openness", False),
    ("I have a rich vocabulary.", "Openness", False),
    ("I have a vivid imagination.", "Openness", False),
    ("I am quick to understand things.", "Openness", False),
    ("I spend time reflecting on things.", "Openness", False),
    ("I am not interested in abstract ideas.", "Openness", True),
    ("I do not like art.", "Openness", True),
    ("I have difficulty understanding abstract ideas.", "Openness", True),
    ("I have a lot of artistic interests.", "Openness", False),
    
    ("I am easily disturbed.", "Neuroticism", False),
    ("I get upset easily.", "Neuroticism", False),
    ("I change my mood a lot.", "Neuroticism", False),
    ("I get nervous easily.", "Neuroticism", False),
    ("I worry about things.", "Neuroticism", False),
    ("I often feel blue.", "Neuroticism", False),
    ("I am relaxed most of the time.", "Neuroticism", True),
    ("I rarely feel depressed.", "Neuroticism", True),
    ("I am easily embarrassed.", "Neuroticism", False),
    
    ("I feel little concern for others.", "Agreeableness", True),
    ("I am not interested in other people's problems.", "Agreeableness", True),
    ("I insult people.", "Agreeableness", True),
    ("I sympathize with others' feelings.", "Agreeableness", False),
    ("I take time out for others.", "Agreeableness", False),
    ("I am not really interested in others.", "Agreeableness", True),
    ("I feel others' emotions.", "Agreeableness", False),
    ("I make people feel at ease.", "Agreeableness", False),
    
    ("I am often preoccupied with details.", "Conscientiousness", False),
    ("I follow a schedule.", "Conscientiousness", False),
    ("I am exacting in my work.", "Conscientiousness", False),
    ("I am always prepared.", "Conscientiousness", False),
    ("I leave my belongings around.", "Conscientiousness", True),
    ("I pay attention to details.", "Conscientiousness", False),
    ("I make plans and stick to them.", "Conscientiousness", False),
    ("I get chores done right away.", "Conscientiousness", False),
]

# ÊòæÁ§∫Ë°®Âçï
st.title("üîç Big Five Personality Test + Career Recommender")

st.markdown("ËØ∑Ê†πÊçÆËá™Â∑±ÁöÑÁúüÂÆûÊÑüÂèóÈÄâÊã©‰∏ãÂàóÊèèËø∞ÁöÑÈÄÇÁî®Á®ãÂ∫¶Ôºö**1ÔºàÈùûÂ∏∏‰∏çÂêåÊÑèÔºâÂà∞ 5ÔºàÈùûÂ∏∏ÂêåÊÑèÔºâ**")

response_dict = {}
for i, (q, _, _) in enumerate(items):
    response = st.slider(f"{i+1}. {q}", 1, 5, 3)
    response_dict[i] = response

if st.button("üéØ Êèê‰∫§Âπ∂Êé®ËçêËÅå‰∏ö"):
    # ÂàÜÁ±ªÈóÆÈ¢ò -> ËÆ°ÁÆóÊØè‰∏™Áª¥Â∫¶ÂæóÂàÜ
    trait_scores = {"Extraversion": [], "Openness": [], "Neuroticism": [], "Agreeableness": [], "Conscientiousness": []}
    for i, (q, trait, reverse) in enumerate(items):
        score = 6 - response_dict[i] if reverse else response_dict[i]
        trait_scores[trait].append(score)
    
    big5_vector = [np.mean(trait_scores[trait]) for trait in ["Neuroticism", "Extraversion", "Openness", "Agreeableness", "Conscientiousness"]]

    # Ê†áÂáÜÂåñ + Ê®°ÂûãÈ¢ÑÊµã
    scaled_input = scaler.transform([big5_vector])
    with torch.no_grad():
        input_tensor = torch.tensor(scaled_input, dtype=torch.float32)
        logits = model(input_tensor).numpy().flatten()
        scores = similarity_matrix @ logits

        top_indices = np.argsort(scores)[-10:][::-1]
        st.subheader("üß† Êé®ËçêËÅå‰∏ö Top-10")
        for i in top_indices:
            st.write(f"{job_codes[i]} - {job_names[i]}  ÔºàÁõ∏‰ººÂ∫¶ÂæóÂàÜÔºö{scores[i]:.3f}Ôºâ")
```
